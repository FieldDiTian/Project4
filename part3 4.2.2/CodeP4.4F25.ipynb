```python

# define neural network model
#In the first layer, the input shape is provided, which is 3 in our case. 
#As seen below, we have included three dense layers. 
#The activation function we have chosen is elu, which stands for exponential linear unit. .

from keras import backend as K
#initialize weights with values between -0.2 and 0.5
initializer = keras.initializers.RandomUniform(minval= -0.2, maxval=0.5)

model = keras.Sequential([
    keras.layers.Dense(13, activation=K.elu, input_shape=[3],  kernel_initializer=initializer),
    keras.layers.Dense(26, activation=K.elu,  kernel_initializer=initializer),
    keras.layers.Dense(13, activation=K.elu, kernel_initializer=initializer), 
    keras.layers.Dense(num_classes, activation='softmax')
  ])
#ADD OUTPUT LAYER, REMOVE DROPOUTS FOR FIRST PART, ADD LATER
```

```python


#COMPLIE THE MODEL
'''After the model is created, you compile it using the Adam optimizer, one of the
most popular optimization algorithms. 

Additionally, you specify the loss type which is categorical cross entropy which
is used for multi-class classification.  Lastly, you specify the metrics as accuracy which you want to
analyze while the model is training.'''
#COMPLETE THE model.compile STATEMENT BELOW
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(), metrics=['accuracy'])


#SUMMARIZE MODEL
'''To visualize the layers that created in the above step, use the summary function.
This will show some parameters (weights and biases) in each layer and also the total parameters in your model.'''
model.summary()
```

```python

#TRAIN THE MODEL


# Add an early stopping callback
es = keras.callbacks.EarlyStopping(
    monitor='loss', 
    mode='min', 
    patience = 2000, 
    restore_best_weights = True, 
    verbose=1)
# Add a checkpoint where loss is minimum, and save that model
mc = keras.callbacks.ModelCheckpoint('best_model.SB', monitor='loss', 
                     mode='min',  verbose=1, save_best_only=True)

#COMPLETE THE model.fit STATEMENT BELOW
model_train = model.fit(train_X, train_label, epochs=2000, verbose=1, validation_data=(valid_X, valid_label))

loss_hist = model_train.history['loss']
#The above line will return a dictionary, access it's info like this:
best_epoch = np.argmin(model_train.history['loss']) + 1
print ('best epoch = ', best_epoch)
print('smallest loss =', np.min(loss_hist))
```

## Task (j) - Add Dropout Layers

```python
# Define model with Dropout(0.25) after hidden layers
model_dropout = keras.Sequential([
    keras.layers.Dense(13, activation=K.elu, input_shape=[3], kernel_initializer=initializer),
    keras.layers.Dense(26, activation=K.elu, kernel_initializer=initializer),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(13, activation=K.elu, kernel_initializer=initializer),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(num_classes, activation='softmax')
])

model_dropout.compile(loss=keras.losses.categorical_crossentropy,
                      optimizer=keras.optimizers.Adam(), metrics=['accuracy'])
model_dropout.summary()
```

```python
# Train Dropout model
es_dropout = keras.callbacks.EarlyStopping(monitor='loss', mode='min', patience=2000, 
                                            restore_best_weights=True, verbose=1)
mc_dropout = keras.callbacks.ModelCheckpoint('best_model_dropout.keras', monitor='loss', 
                                             mode='min', verbose=1, save_best_only=True)

model_dropout_train = model_dropout.fit(train_X, train_label, epochs=2000, verbose=1, 
                                        validation_data=(valid_X, valid_label))

loss_hist_dropout = model_dropout_train.history['loss']
best_epoch_dropout = np.argmin(loss_hist_dropout) + 1
print(f'Best epoch: {best_epoch_dropout}, Smallest loss: {np.min(loss_hist_dropout):.6f}')
```

```python
# Evaluate Dropout model
print('=== Dropout Model Evaluation ===\n')

train_eval_dropout = model_dropout.evaluate(train_X, train_label, verbose=0)
print(f'Train loss: {train_eval_dropout[0]:.6f}')
print(f'Train accuracy: {train_eval_dropout[1]:.4f}\n')

test_eval_dropout = model_dropout.evaluate(valid_X, valid_label, verbose=0)
print(f'Validation loss: {test_eval_dropout[0]:.6f}')
print(f'Validation accuracy: {test_eval_dropout[1]:.4f}\n')

# Loss ratio comparison
loss_ratio_orig = 0.7179 / 0.0014
loss_ratio_dropout = test_eval_dropout[0] / train_eval_dropout[0]
print(f'Loss ratio (val/train):')
print(f'  Original model: {loss_ratio_orig:.1f}')
print(f'  Dropout model: {loss_ratio_dropout:.1f}')
print(f'\n>>> Copy results to log file <<<')
```

```python
# Task 2.1 power predictions (from CodeP4.2F25.ipynb)
# Format: (Ta, ID, RL): [W_M0, W_M1, W_M2]
task21_power_predictions = {
    (10.0, 200, 50.0): [47.30, 139.53, 43.19],
    (20.0, 200, 130.0): [-16.52, 88.62, 118.05],
    (10.0, 500, 40.0): [54.41, 219.32, 155.47],
    (20.0, 500, 80.0): [30.24, 123.38, 117.66],
    (20.0, 700, 30.0): [82.19, 344.30, 227.63],
    (20.0, 700, 55.0): [47.73, 160.93, 173.26],
    (10.0, 1000, 12.0): [204.31, 922.55, 575.11],
    (20.0, 1000, 25.0): [118.56, 488.56, 815.65],
    (20.0, 1000, 39.0): [55.82, 232.39, 567.44],
}
```

```python
# Predict Mmax using Task 4.2.2 model
test_conditions = [
    [10.0, 200, 50.0], [20.0, 200, 130.0], [10.0, 500, 40.0],
    [20.0, 500, 80.0], [20.0, 700, 30.0], [20.0, 700, 55.0],
    [10.0, 1000, 12.0], [20.0, 1000, 25.0], [20.0, 1000, 39.0]
]

results = []
for Ta, ID, RL in test_conditions:
    input_norm = np.array([[Ta/Tamed, ID/IDmed, RL/RLmed]])
    Mmax_pred = np.argmax(model.predict(input_norm, verbose=0)[0])
    powers = task21_power_predictions[(Ta, ID, RL)]
    actual_Mmax = np.argmax(powers)
    
    results.append({
        'Ta': Ta, 'ID': ID, 'RL': RL,
        'Mmax_pred': Mmax_pred,
        'W': powers,
        'actual_Mmax': actual_Mmax,
        'correct': Mmax_pred == actual_Mmax
    })

print('Prediction Results:')
print('='*100)
for i, r in enumerate(results, 1):
    check = 'âœ“' if r['correct'] else 'âœ—'
    print(f"{i}. Ta={r['Ta']:.0f}Â°C ID={r['ID']:.0f} RL={r['RL']:.0f} | "
          f"Pred={r['Mmax_pred']} Actual={r['actual_Mmax']} | "
          f"W=[{r['W'][0]:.1f}, {r['W'][1]:.1f}, {r['W'][2]:.1f}] {check}")
print('='*100)
```

```python
# Calculate accuracy
correct = sum(1 for r in results if r['correct'])
total = len(results)
accuracy = correct / total * 100

print(f'\nAccuracy: {correct}/{total} = {accuracy:.2f}%\n')
print('>>> Copy results to log file <<<')
```
